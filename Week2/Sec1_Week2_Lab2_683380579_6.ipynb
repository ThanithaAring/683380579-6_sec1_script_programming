**Lab 2.1: Number Classifier**

○ Objective: Students will write a script to classify a user-provided integer as positive,
negative, or zero, and also check if it&#39;s even or odd.

○ Tasks (Common for Local &amp; Colab):
1. Create a new Python file (lab2_1.py) or Colab code cell.
2. Prompt the user to enter an integer using input().
3. Convert the input to an integer.
4. Use if-elif-else to determine if the number is positive, negative, or zero. Print the
classification.
5. Use an if-else statement to determine if the number is even or odd. Print the result.
6. Challenge: Combine the checks using logical operators. For example, print &quot;The
number is positive and even.&quot; or &quot;The number is negative and odd.&quot;

```python
user_input = input("Please input integer: ")

number = int(user_input)

if number > 0:
    print(f"{number} is positive")
elif number < 0:
    print(f"{number} is negative")
else:
    print(f"{number} is zero")

if number % 2 == 0:
    print(f"{number} is even")
else:
    print(f"{number} is odd")

if number > 0 and number % 2 == 0:
    print(f"{number} is positive and even")
elif number > 0 and number % 2 != 0:
    print(f"{number} is positive and odd")
elif number < 0 and number % 2 == 0:
    print(f"{number} is negative and even")
elif number < 0 and number % 2 != 0:
    print(f"{number} is negative and odd")
else:
    print(f"{number} is zero.")
```

# Block A: Evaluation Check


```python
temp = 25
is_raining = True

if temp > 20 and not is_raining:
    print("Perfect day for a picnic!")
elif temp <= 20 or is_raining:
    print("Stay indoors or bring an umbrella.")
else:
    print("Unsure about the weather.")
```

เช็คเงื่อนไข if แรกคือ temp > 20 and not is_raining

not is_raining คือ not True เท่ากับว่า False
True and False เป็น False
เงื่อนไข if เป็น False ก็จะข้ามไปเช็ค elif


เช็คเงื่อนไข elif คือ temp <= 20 or is_raining

False or True คือ True

**ผลลัพธ์:**

Stay indoors or bring an umbrella.

# Block B: Vectorized Filter


```python
phrase = "Python 2026 Engine"
consonants = {c.lower() for c in phrase if c.isalpha() and c.lower() not in 'aeiou'}
print(f"Unique consonants: {consonants}")
```

**ลำดับการทำงาน:**

วนแต่ละตัวอักษรใน "Python 2026 Engine"

c.isalpha() คือต้องเป็นตัวอักษร (ตัด space และตัวเลข 2026 ทิ้ง)

c.lower() not in 'aeiou' คือต้องไม่ใช่สระ

1.ตัวอักษรทั้งหมดมี P,y,t,h,o,n, ,2,0,2,6, ,E,n,g,i,n,e

2.คัดเฉพาะตัวอักษรแล้วแปลงเป็นตัวพิมพ์เล็กมี p,y,t,h,o,n,e,n,g,i,n,e

3.ตัดสระออกก็จะเหลือ p,y,t,h,n,n,g,n

4.เพราะเป็น set จะเก็บเฉพาะค่าที่ไม่ซ้ำก็จะเหลือ {p, y, t, h, n, g}

# **Activity 2.2: Choose Your Own Adventure Architectures**

● Format: Small Groups (2–3 students).

● Task: Outline a short interactive narrative mapping out 2–3 explicit structural game
decisions.

● Deliverable: Build an algorithmic pseudocode diagram documenting the conditional flow
mechanics before presenting your layout maps to the class.

```python
print("Welcome to the Three Little Pigs game!")

material = input("A big bad wolf is coming to blow down your house! \nChoose your building material 1=Straw 2=Wood 3=Brick: ")
material = int(material)

if material == 1:
    # Straw
    print("\nThe straw house was built fast, but it looks flimsy")
    print("The wolf huffs just once and the house blows away instantly!")
    action = input("Do you repair the house or run away? (repair/run): ")

    if action == "repair":
        print("\nYou rush to repair the house, but the wolf huffs again!")
        print("The house finally collapses. You die!")
    elif action == "run":
        print("\nYou decide to run just in time!")

        house = input("Which house are you running to (2=Wood house/3=Brick house): ")
        if house == "2":
            print("The wooden house took a while to fall, but it still collapsed. You die!")
        elif house == "3":
            print("You escape to your youngest sibling's brick house - You survive!")
        else:
            print("Invalid answer. There's no other house, you die!")
    else:
        print("Invalid answer. Please type 'repair' or 'run' only.")

elif material == 2:
    # Wood
    print("\nThe wooden house took a while, and looks fairly sturdy.")
    print("The wolf huffs harder! The house shakes but doesn't collapse yet.")
    action = input("Do you repair the house or run away? (repair/run): ")

    if action == "repair":
        print("\nYou rush to repair the house, but the wolf huffs again!")
        print("The house finally collapses. You die!")
    elif action == "run":
        print("\nYou decide to run just in time!")

        house = input("Which house are you running to (1=Straw house/3=Brick house): ")
        if house == "1":
            print("The straw house is already destroyed, there's nowhere to hide. You die!")
        elif house == "3":
            print("You escape to your youngest sibling's brick house - You survive!")
        else:
            print("Invalid answer. There's no other house, you die!")
    else:
        print("Invalid answer. Please type 'repair' or 'run' only.")

elif material == 3:
    # Brick
    print("\nThe brick house took the longest to build, but looks very strong.")
    print("No matter how hard the wolf huffs, the house doesn't budge!")
    print("The wolf gives up, exhausted.")
    print("Congratulations! The brick house is the strongest. You survive!")

else:
    print("Please choose only 1, 2, or 3.")
```
