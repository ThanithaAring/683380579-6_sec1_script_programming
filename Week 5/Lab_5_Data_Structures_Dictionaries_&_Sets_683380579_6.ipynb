```python
contacts = {}
while True:
    print("\n=== Contact Book Manager ===")
    print("1. Add/Update Contact")
    print("2. View Contact Details")
    print("3. List All Contacts")
    print("4. Delete Contact")
    print("5. Exit")

    choice = input("Enter choice (1-5): ")

    if choice == '1':
        name = input("Enter contact name: ").strip().title()
        phone = input("Enter phone number: ")
        email = input("Enter email: ")

        contacts[name] = {
            "phone": phone,
            "email": email
        }
        print(f"Contact '{name}' saved successfully!")

    elif choice == '2':
        name = input("Enter contact name to view: ").strip().title()

        info = contacts.get(name)

        if info:
            print(f"\nName: {name}")
            print(f"Phone: {info['phone']}")
            print(f"Email: {info['email']}")
        else:
            print("Contact not found.")

    elif choice == '3':
        if not contacts:
            print("No contacts available.")
        else:
            print("\n--- All Contacts ---")
            for name, details in contacts.items():
                print(f"Name: {name} | Phone: {details['phone']} | Email: {details['email']}")

    elif choice == '4':
        name = input("Enter contact name to delete: ").strip().title()

        removed_contact = contacts.pop(name, None)

        if removed_contact:
            print(f"Contact '{name}' deleted successfully!")
        else:
            print("Contact not found.")

    elif choice == '5':
        print("Exiting Contact Book Manager. Goodbye!")
        break

    else:
        print("Invalid choice! Please enter a number from 1 to 5.")
```

### กิจกรรมที่ 5.2: โปรแกรมนับความถี่ของคำ (Word Frequency Counter) โดยใช้ Dictionary

* **รูปแบบ:** โจทย์ท้าทายการเขียนโค้ดรายบุคคล (Individual coding challenge)
* **คำชี้แจง:** ให้นักเรียนเขียนโปรแกรมรับข้อความ (ประโยค) จากผู้ใช้ แล้วนับจำนวนความถี่ของแต่ละคำโดยใช้ Dictionary
* **ขั้นตอนการทำงาน (Steps):**
1. รับข้อความประโยคจากผู้ใช้
2. แปลงข้อความให้เป็นตัวพิมพ์เล็กทั้งหมด: `sentence.lower()`
3. แยกประโยคออกเป็นคำๆ: `words = sentence.split()`
4. กำหนด Dictionary เปล่าขึ้นมา: `word_counts = {}`
5. วนลูปอ่านคำแต่ละคำในลิสต์ `words`
6. ภายในลูป ใช้ `if word in word_counts:` เพื่อเช็กว่าคำนั้นมีอยู่ใน Key ของ Dictionary แล้วหรือยัง
7. ถ้า **มีแล้ว** ให้เพิ่มค่าจำนวนนับขึ้นอีก 1: `word_counts[word] += 1`
8. ถ้า **ยังไม่มี** ให้เพิ่มคำนั้นเข้า Dictionary พร้อมกำหนดค่าเริ่มต้นเป็น 1: `word_counts[word] = 1`
9. แสดงผลลัพธ์ของ Dictionary `word_counts`


* **ความท้าทายเพิ่มเติม (Challenge):** ให้จัดการกับเครื่องหมายวรรคตอนด้วย *(เช่น ลบเครื่องหมายจุลภาค `,` หรือจุด `.` ออกก่อนนำคำไปนับ)* **คำใบ้:** ใช้คำสั่ง `word.strip('.,!?"')`
* **วัตถุประสงค์:** เพื่อประยุกต์ใช้การทำงานของ Dictionary กับโจทย์การประมวลผลข้อความจริง แสดงให้เห็นถึงประสิทธิภาพในการนับและการค้นหาข้อมูล

**กิจกรรม 5.1**

1.)

-คลังไอเทมของผู้เล่น (Inventory):

ใช้ชื่อไอเทมเป็น key และจำนวนที่มีเป็น value เช่น inventory = {"potion": 5, "sword": 1, "gold_coin": 120}

-ทำให้ค้นหาและอัปเดตจำนวนไอเทมได้รวดเร็วโดยไม่ต้องวนลูปหาทั้ง list
สถานะตัวละคร (Character Stats):

เช่น stats = {"hp": 100, "mana": 50, "attack": 25, "defense": 10} การเข้าถึงค่าแต่ละสถานะทำได้ทันทีผ่านชื่อ key ที่สื่อความหมาย

เหตุผลที่เหมาะสม: dictionary ให้การเข้าถึงข้อมูลแบบ O(1) ผ่าน key ที่มีความหมายชัดเจน ต่างจาก list ที่ต้องจำตำแหน่ง index

2.)

-ใช้รหัสสินค้า (product ID) เป็น key ใน dictionary ใหญ่ที่เก็บสินค้าทั้งหมด เช่น products = {"P10023": {...}, "P10024": {...}}

-ทำให้ค้นหาสินค้าตามรหัสได้ทันทีโดยไม่ต้องค้นหาทีละรายการ
เหมาะสำหรับระบบที่ต้อง lookup ข้อมูลสินค้าบ่อย ๆ เช่น ตะกร้าสินค้า หรือหน้าตรวจสอบสต็อก

3.)

-บทบาทผู้ใช้ (Roles):

user_roles = {"admin", "editor"} การตรวจสอบสิทธิ์ทำได้ง่ายด้วย "admin" in user_roles และไม่มีทางที่ role จะซ้ำกันในระบบเดียวโดยไม่ตั้งใจ

-แฮชแท็ก (Tags/Hashtags):

โพสต์หนึ่งอาจมีแท็กซ้ำที่ผู้ใช้พิมพ์ผิดพลาด set จะช่วยลบแท็กซ้ำโดยอัตโนมัติ เช่น tags = set(["python", "coding", "python"]) → {"python", "coding"}
การรวมสิทธิ์จากหลายกลุ่ม (union) หรือหาสิทธิ์ที่ผู้ใช้สองคนมีร่วมกัน (intersection) ก็ทำได้ง่ายด้วยการดำเนินการทางเซต

4.)

IP address ที่ถูกบล็อก: blocked_ips = {"192.168.1.10", "203.0.113.5", ...} เมื่อมีการเชื่อมต่อเข้ามา ระบบตรวจสอบด้วย if ip in blocked_ips: ซึ่งเร็วกว่าการวนลูปใน list มาก โดยเฉพาะเมื่อมีข้อมูลนับล้านรายการ
ผู้เยี่ยมชมที่ไม่ซ้ำกัน (Unique Visitors): เก็บ user ID ของผู้เข้าชมเว็บไซต์ในหนึ่งวันด้วย set เพื่อนับจำนวนผู้เยี่ยมชมที่ไม่ซ้ำกันได้ทันที โดยไม่ต้องเขียนโค้ดตรวจสอบค่าซ้ำเอง

เหตุผลที่เหมาะสม: การตรวจสอบสมาชิก (in) ใน set ใช้เวลาเฉลี่ยคงที่ (average O(1)) ต่างจาก list ที่ต้องตรวจสอบทีละตัว (O(n)) จึงเหมาะกับข้อมูลขนาดใหญ่ที่ไม่สนใจลำดับและไม่มีค่าซ้ำ

```python
# Activity 5.2

sentence = input("Enter a sentence: ")

sentence = sentence.lower()

words = sentence.split()

word_counts = {}

for word in words:
    word = word.strip('.,!?"\'')

    if word == "":
        continue

    if word in word_counts:
        word_counts[word] += 1
    else:
        word_counts[word] = 1

print("\nWord Frequency")
for word, count in word_counts.items():
    print(f"{word}: {count}")
```
