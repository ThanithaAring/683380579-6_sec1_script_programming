Activity 1.2: "Python Puzzlers" - Identifying Data Types & Output
Format: Individual or pair work, followed by a class review.
Instructions: Provide a set of small Python code snippets (on a handout or projected). For each snippet, students must:
Predict the output.
Identify the data type of specific variables after certain operations.
Explain why they chose that output/data type.


```python
# Puzzler 1
x = 10
y = 3.5
z = x + y
print(z)
print(type(z))
```

**ได้ผลลัพธ์:** 13.5
<class 'float'>

**ทำไมถึงเป็นtype float :** เพราะ x = 10 เป็น int และ y = 3.5 เป็น float เมื่อบวกกันจะได้ผลลัพธ์เป็น float (int + float = float)

```python
# Puzzler 2
name = "Alice"
age = "30" # Note: This is a string
message = name + " is " + age + " years old."
print(message)
print(type(age))
```

**ได้ผลลัพธ์:** Alice is 30 years old.
<class 'str'>

**ทำไมถึงเป็นtype :** เพราะ age = "30" ใส่เครื่องหมาย " " ทำให้ age เป็น string ไม่ใช่ int และการใช้ + ระหว่าง string คือการต่อข้อความ

```python
# Puzzler 3
is_sunny = True
is_raining = False
print(is_sunny and is_raining)
print(type(is_sunny))
```

**ได้ผลลัพธ์:** False
<class 'bool'>

**ทำไมถึงเป็นtype :** เพราะใช้ and ซึ่งจะได้Trueก็ต่อเมื่อ ทั้งสองค่าเป็นTrue และ type(is_sunny) คือ bool เพราะ True/False เป็น Boolean data type
