**Lab 4.1: Student Management System (List Operations)**

Objective: Students will build a simple program to manage a list of student
names, demonstrating various list operations.


```python
student_names = []

print("\n[1] Add Students")
for i in range(3):
    name = input(f"Enter student name #{i + 1}: ")
    student_names.append(name)

print(f"Current list: {student_names}")

print("\n[2] Find a Student")
search_name = input("Enter a name to search for: ")

if search_name in student_names:
    position = student_names.index(search_name)
    print(f"Found '{search_name}' at index {position}")
else:
    print(f"'{search_name}' not found in the list")

print("\n[3] Remove a Student")
remove_name = input("Enter a name to remove: ")

if remove_name in student_names:
    student_names.remove(remove_name)
    print(f"'{remove_name}' has been removed")
else:
    print(f"'{remove_name}' not found, nothing removed")

print(f"Updated list: {student_names}")

print("\n[4] Sort Students")
student_names.sort()
print(f"Sorted list: {student_names}")

print("\n[5] Count Students")
print(f"Total students: {len(student_names)}")

def add_student(names_list):
    new_name = input("Enter the name of the student to add: ")
    names_list.append(new_name)
    print(f"'{new_name}' has been added.")
    print(f"Current list: {names_list}")


def remove_student(names_list):
    name = input("Enter the name of the student to remove: ")
    if name in names_list:
        names_list.remove(name)
        print(f"'{name}' has been removed.")
    else:
        print(f"'{name}' not found, nothing removed.")
    print(f"Current list: {names_list}")


def find_student(names_list):
    name = input("Enter a name to search for: ")
    if name in names_list:
        position = names_list.index(name)
        print(f"Found '{name}' at index {position}")
    else:
        print(f"'{name}' not found in the list")


def sort_students(names_list):
    names_list.sort()
    print(f"Sorted list: {names_list}")


def show_menu():
    print("\n===== Student Management Menu =====")
    print("1. Add a student")
    print("2. Remove a student")
    print("3. Find a student")
    print("4. Sort students")
    print("5. View all students")
    print("6. Exit")


def run_menu(names_list):
    while True:
        show_menu()
        choice = input("Choose an option (1-6): ")

        if choice == "1":
            add_student(names_list)
        elif choice == "2":
            remove_student(names_list)
        elif choice == "3":
            find_student(names_list)
        elif choice == "4":
            sort_students(names_list)
        elif choice == "5":
            print(f"Students ({len(names_list)}): {names_list}")
        elif choice == "6":
            print("Exiting program. Goodbye!")
            break
        else:
            print("Invalid choice, please enter a number from 1 to 6.")


if __name__ == "__main__":
    run_menu(student_names)
```

**Lab 4.2: Geometric Calculations (Tuples & Immutability)**

Objective: Students will use tuples to represent 2D points and calculate the distance between them, highlighting tuple immutability.


```python
import math
point1 = (3, 4)
point2 = (6, 8)

print("\n[1] Access Coordinates")
print(f"Point 1: x = {point1[0]}, y = {point1[1]}")
print(f"Point 2: x = {point2[0]}, y = {point2[1]}")

print("\n[2] Attempt Modification (Immutability Demonstration)")
try:
    point1[0] = 5
except TypeError as e:
    print(f"Error trying to modify tuple: {e}")

print("\n[3] Calculate Distance")
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2)
print(f"Distance between point1 {point1} and point2 {point2}: {distance}")

print("\n[4] Challenge: Enter Your Own Points")

x1 = float(input("Enter x for point A: "))
y1 = float(input("Enter y for point A: "))
point_a = (x1, y1)

x2 = float(input("Enter x for point B: "))
y2 = float(input("Enter y for point B: "))
point_b = (x2, y2)

user_distance = math.sqrt((point_b[0] - point_a[0]) ** 2 + (point_b[1] - point_a[1]) ** 2)

print(f"Distance between point A {point_a} and point B {point_b}: {user_distance}")
```

**Activity 4.2:** Interactive To-Do List (using List Operations)
Format: Individual coding challenge.

**Instructions:** Students will implement a very basic text-based to-do list manager using what they've learned about lists.

**Features to include (minimum):**
An empty list todo_list initially.

A while loop that presents a menu:

Add a task

View tasks

Mark task as complete (remove by index or name)

Exit

Use input() to get user choices.

Use list methods (append(), remove() or pop(), len()) to manage the todo_list.

**Purpose:** Practical application of list operations in a mini-project context, integrating loops and conditionals.


```python
todo_list = []

def show_menu():
    print("\n===== To-Do List Menu =====")
    print("1. Add a task")
    print("2. View tasks")
    print("3. Mark task as complete (remove)")
    print("4. Exit")

def add_task(tasks):
    task = input("Enter a new task: ")
    tasks.append(task)
    print(f"Task added: '{task}'")

def view_tasks(tasks):
    if len(tasks) == 0:
        print("Your to-do list is empty.")
    else:
        print("\n--- Your Tasks ---")
        for i, task in enumerate(tasks):
            print(f"{i + 1}. {task}")

def complete_task(tasks):
    if len(tasks) == 0:
        print("No tasks to complete.")
        return

    view_tasks(tasks)
    choice = input("Enter the task number OR the task name to mark complete: ")

    if choice.isdigit():
        index = int(choice) - 1
        if 0 <= index < len(tasks):
            removed = tasks.pop(index)
            print(f"Completed and removed: '{removed}'")
        else:
            print("Invalid task number.")
    else:
        if choice in tasks:
            tasks.remove(choice)
            print(f"Completed and removed: '{choice}'")
        else:
            print(f"Task '{choice}' not found.")

def run_todo_app():
    while True:
        show_menu()
        choice = input("Choose an option (1-4): ")

        if choice == "1":
            add_task(todo_list)
        elif choice == "2":
            view_tasks(todo_list)
        elif choice == "3":
            complete_task(todo_list)
        elif choice == "4":
            print("Exiting To-Do List. Goodbye! >:)")
            break
        else:
            print("Invalid choice, please enter a number from 1 to 4.")

if __name__ == "__main__":
    run_todo_app()
```
